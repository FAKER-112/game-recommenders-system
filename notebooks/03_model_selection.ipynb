```python
from tensorflow.keras.models import load_model

model = load_model("../artifacts/models/mf_model.h5", compile=False)
from tensorflow.keras.utils import plot_model

plot_model(model, to_file='model_structure.png', show_shapes=True, show_layer_names=True)
```

```python
import tensorflow as tf

logdir = "logs/model_graph"
writer = tf.summary.create_file_writer(logdir)

tf.summary.trace_on(graph=True, profiler=False)

# Run one forward pass to trace the graph:
dummy_input = tf.random.uniform((1, *model.input_shape[1:]))
model(dummy_input)

with writer.as_default():
    tf.summary.trace_export(name="model_trace", step=0)
```

```python
import tensorflow as tf

logdir = "logs/model_graph"
writer = tf.summary.create_file_writer(logdir)

tf.summary.trace_on(graph=True, profiler=False)

# Run one forward pass to trace the graph:
dummy_input = tf.random.uniform((1, model.input_shape[1:]))
model(dummy_input)

with writer.as_default():
    tf.summary.trace_export(name="model_trace", step=0)
```

```python
dummy_inputs = []
for inp in model.inputs:
    shape = [1 if d is None else int(d) for d in inp.shape[1:]]
    dummy_inputs.append(tf.random.uniform([1] + shape))

tf.summary.trace_on(graph=True, profiler=False)
model(dummy_inputs)

with writer.as_default():
    tf.summary.trace_export(name="model_trace", step=0)
```
